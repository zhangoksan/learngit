### 2.2.2 完善原始折线图（辅助显示层）

```python
# 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围15-18度
import random
# 1. 准备数据 x , y
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]

# 2. 创建画布
plt.figure(figsize=(25,8), dpi=80)

# 3. 绘制图像
plt.plot(x, y_shanghai)

# 修改x, y刻度
# 准备x的刻度说明
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[: :5], x_label[: :5])
plt.yticks(range(0, 40, 5))

# 添加网格显示
plt.grid(True, linestyle= "--", alpha=0.5)

# 添加描述信息
plt.xlabel("时间变化")
plt.ylabel("温度")
plt.title("某城市11点到12点每分钟温度变化")

# 4. 显示图像
plt.show()
```
